# Função 1 - O(n)

```python
def soma1(n):
    soma = 0
    for i in range(n + 1):
        soma += i
    
    return soma
```

```python
soma1(10)
```

```python
%timeit soma1(10)
```

# Funções Big-O
 - Constante
 - Log(n)
 - Linear
 - n log(n) Log Linear
 - n^2
 - n^3
 - 2^n (Exponencial)

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt 

n = np.linspace(1,10,100) # composto por 100 numeros entre  1 e 10

len(n)
```

```python
labels = ['Constant'      ,'Logarithmic','Linear','Log linear' ,'Quadratic','Cubic','Exponential']
big_o =  [np.ones(n.shape),np.log(n)    ,n       ,n * np.log(n),n**2       , n**3  ,2**n         ]
```

```python
plt.figure(figsize=(10,8))
plt.ylim(0,100)
for i in range(len(big_o)):
    plt.plot(n, big_o[i], label = labels[i])
plt.legend()
plt.ylabel('Runtime')
plt.xlabel('n')
```

# Exemplos Big-O

## Constant

```python
lista = [1,2,3,4,5]
```

```python
def constant(n):
    print(n[0])
    print(n[1])
```

```python
constant(lista)
```

## Linear

```python
def linear(n):
    for i in n:
        print(i)
```

```python
linear(lista)
```

## Quadratica

```python
def quadratic(n):
    for i in n:
        for j in n:
            print(i,j)
        print('-----')
```

```python
quadratic(lista)
```

## Combination

```python
def combination(n):
    # O(1)
    print(n[0])
    # O(5)
    for i in range(5):
        print('test', i)
    # O(n)
    for i in n:
        print(i)
    # O(n)
    for i in n:
        print(i)
    # O(3)
    print('Python')
    print('Python')
    print('Python')
```

```python
combination(lista)
```
