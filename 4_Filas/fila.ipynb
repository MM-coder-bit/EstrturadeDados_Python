```python
import numpy as np
```

```python
# Implementação de uma Fila Circular em Python
class FilaCircular:

    def __init__(self, capacidade):
        # Inicializa a fila circular com uma capacidade específica
        self.capacidade = capacidade
        # Inicializa o início da fila como 0
        self.inicio = 0
        # Inicializa o final da fila como -1
        self.final = -1
        # Inicializa o número de elementos na fila como 0
        self.numero_elementos = 0
        # Inicializa um array para armazenar os valores da fila
        self.valores = np.empty(self.capacidade, dtype=int)

    def __fila_vazia(self):
        # Verifica se a fila está vazia
        return self.numero_elementos == 0

    def __fila_cheia(self):
        # Verifica se a fila está cheia
        return self.numero_elementos == self.capacidade

    def enfileirar(self, valor):
        # Adiciona um elemento ao final da fila circular
        if self.__fila_cheia():
            print('A fila está cheia')
            return

        # Atualiza o índice do final da fila circular, considerando a circularidade
        if self.final == self.capacidade - 1:
            self.final = -1
        self.final += 1
        self.valores[self.final] = valor
        self.numero_elementos += 1

    def desenfileirar(self):
        # Remove e retorna o elemento do início da fila circular
        if self.__fila_vazia():
            print('A fila já está vazia')
            return

        # Armazena temporariamente o valor a ser removido
        temp = self.valores[self.inicio]

        # Atualiza o índice do início da fila circular, considerando a circularidade
        self.inicio += 1
        if self.inicio == self.capacidade - 1:
            self.inicio = 0

        # Atualiza o número de elementos na fila
        self.numero_elementos -= 1
        return temp

    def primeiro(self):
        # Retorna o valor no início da fila circular sem removê-lo
        if self.__fila_vazia():
            return -1
        return self.valores[self.inicio]
```

```python
f = FilaCircular(5)
# Veridicar o primeiro elemento da fila. Se não houver, retorna -1.
print(5 * '-', 'Fila Vazia', 5 * '-')
print(f.primeiro())

# enfilera
f.enfileirar(1)
f.enfileirar(2)
f.enfileirar(3)
f.enfileirar(4)
f.enfileirar(5)
print(5 * '-', 'O primeiro da Fila', 5 * '-')
# a sequencia esta [1,2,3,4,5]
print(f.primeiro())

# Validando que a fila esta cheia
print(5 * '-', 'Limite da fila Ultrapassado', 5 * '-')
f.enfileirar(6)

# desinfilerar
print(5 * '-', 'Desinfilerado 2 elementos da fila', 5 * '-')
f.desenfileirar()
f.desenfileirar()
# agora a sequencia eh [3, 4, 5]
print(f.primeiro())

# enfilerando novos dados
print(5 * '-', 'Enfilerado 2 elementos da fila', 5 * '-')
f.enfileirar(6)
f.enfileirar(7)
# agora a sequencia eh [3, 4, 5, 6, 7]
print(f.primeiro())

# apresenta todos dados do vetor
print(5 * '-', 'Apresenta todos os dados da fila', 5 * '-')
print(f.valores)

# verifica a posição do incio e do final da fila
print(5 * '-', 'Apresenta as posições inicio e fim da fila', 5 * '-')
print(f.inicio, f.final)

#verifica os valores do inicio e final da fila
print(5 * '-', 'Apresenta o valor inicio e fim da fila', 5 * '-')
print(f.valores[f.inicio])
print(f.valores[f.final])
```
